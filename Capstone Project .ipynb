```python
model=smf.glm("RETURNED_2ND_YR~STDNT_AGE+STDNT_GENDER+STDNT_TEST+IN_STATE_FLAG+INTERNATIONAL_STS+STDNT_MINOR_NEW+FIRST_TERM+DISTANCE_FROM_HOME+HIGH_SCHL_GPA+FIRST_TERM_ATTEMPT_HRS+FIRST_TERM_EARNED_HRS+SECOND_TERM_ATTEMPT_HRS+SECOND_TERM_EARNED_HRS+FULL_SCHOLARSHIP+EST_FAM_CONTRIBUTION+UNMET_NEED+STDNT_BACKGROUND_BGD_1+STDNT_BACKGROUND_BGD_2+STDNT_BACKGROUND_BGD_3+STDNT_BACKGROUND_BGD_4+STDNT_BACKGROUND_BGD_5+STDNT_BACKGROUND_BGD_6+STDNT_BACKGROUND_BGD_7+STDNT_BACKGROUND_BGD_8+CORE_COURSE_COMM_1110+CORE_COURSE_ENGL_1101+CORE_COURSE_HIST_2111+CORE_COURSE_MATH_1111+CORE_COURSE_Other+CORE_COURSE_GRADE_1_F_A+CORE_COURSE_GRADE_1_F_B+CORE_COURSE_GRADE_1_F_C+CORE_COURSE_GRADE_1_F_D+CORE_COURSE_GRADE_1_F_F+CORE_COURSE_GRADE_1_F_INCOMPL+CORE_COURSE_GRADE_1_F_NOT_REP+HOUSING_STS_Off_Campus+HOUSING_STS_On_Campus+DEGREE_GROUP_CD_A+DEGREE_GROUP_CD_B+DEGREE_GROUP_CD_V+DEGREE_GROUP_DESC_Associate+DEGREE_GROUP_DESC_Bachelors+DEGREE_GROUP_DESC_Career_Associate",data=trainX, family=sm.families.Binomial()).fit()
```

```python
print(model.summary())
```

```python
#confusion Matrix and AUC
import sklearn.metrics as metrics
```

```python
trainX.drop('RETURNED_2ND_YR',inplace=True,axis=1)
testX.drop('RETURNED_2ND_YR',inplace=True,axis=1)
X.drop("RETURNED_2ND_YR",inplace=True,axis=1)
```

```python
y_true=testY
y_pred=model.predict(testX)
```

```python
y_true=testY
y_pred=model.predict(testX).map(lambda x:1 if x>0.5 else 0)
CM=metrics.confusion_matrix(y_true,y_pred)
print(CM)
```

```python
total=sum(sum(CM))
accuracy=(CM[0,0]+CM[1,1])/total
print('Accuracy : ',accuracy*100)
```

## CONCLUSION

If the university wants to reduce the early attrition rate, it should take following steps-

1.Students should be encouraged to attempt less hours in their first term, whereas attempt more hours in thier second term.

2.University shouldn't select students having more age(above 20).

3.University shouldn't select students having Combined test scores between 800-1200, as they are more likely to go for early attrition.

4.University shouldn't select Students having HIGH_SCHL_GRADES around 3.0, as they are more ikely to leave the univeristy after first year.
